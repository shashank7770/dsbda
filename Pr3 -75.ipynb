```python
df = sns.load_dataset('iris')
```

```python
se = df[df['species'] == 'setosa']
ve = df[df['species'] == 'versicolor']
vi = df[df['species'] == 'virginica']
```

```python
def display_statistics(species_df, species_name):
    print(f"\nStatistics for {species_name}:")
    print(f"Mean:\n{species_df.mean()}")
    print(f"Standard Deviation:\n{species_df.std()}")
    print(f"Percentiles (25%, 50%, 75%):\n{species_df.quantile([0.25, 0.5, 0.75])}")
    print(f"Variance:\n{species_df.var()}")
    print(f"Min:\n{species_df.min()}")
    print(f"Max:\n{species_df.max()}")
    print("-" * 50)
```

```python
display_statistics(se.drop('species', axis=1), 'Iris-setosa')
display_statistics(ve.drop('species', axis=1), 'Iris-versicolor')
display_statistics(vi.drop('species', axis=1), 'Iris-virginica')
```
