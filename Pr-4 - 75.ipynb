```python
import numpy as np
import matplotlib.pyplot as plt

# Select the first feature of x_train for plotting
x_feature = x_train.iloc[:, 0]

# Calculate the best-fit line (degree 1 polynomial)
slope, intercept = np.polyfit(x_feature, y_train, 1)

# Create the best-fit line
best_fit_line = np.poly1d([slope, intercept])

# Plot the scatter plot
plt.scatter(x_feature, y_train, color='blue', label='Data points')

# Plot the best-fit line
plt.plot(x_feature, best_fit_line(x_feature), color='red', label='Best fit line')

# Label the axes and add a title
plt.xlabel('First Feature')
plt.ylabel('Tax')
plt.title('Scatter plot with Best Fit Line')

# Show the legend
plt.legend()

# Display the plot
plt.show()
```
