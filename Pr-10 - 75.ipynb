```python
corr = df.select_dtypes(include='number').corr()
plt.figure(figsize=(10, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

#unique colors for each subplot
colors = ['deep', 'bright', 'dark', 'husl']

def graph(y, color_index):
    sns.boxplot(x="Species", y=y, data=df, palette=colors[color_index])

plt.figure(figsize=(12, 10))
plt.subplot(221)
graph('SepalLengthCm', 0)

plt.subplot(222)
graph('SepalWidthCm', 1)

plt.subplot(223)
graph('PetalLengthCm', 2)

plt.subplot(224)
graph('PetalWidthCm', 3)

plt.tight_layout()
plt.show()
```

```python
sns.boxplot(x='SepalWidthCm', data=df)
plt.show()
sns.boxplot(x='SepalLengthCm', data=df)
plt.show()
sns.boxplot(x='PetalWidthCm', data=df)
plt.show()
sns.boxplot(x='PetalLengthCm', data=df)
plt.show()
```
